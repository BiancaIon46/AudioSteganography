```python
import sounddevice as sd
import numpy as np
from matplotlib import pyplot as plt
```

```python
duration = 85 # seconds
fs = 44100
myrecording = sd.rec(int(duration * fs), samplerate = fs, channels = 2)
sd.wait()

print("gata inregistrare")

myrec = myrecording.mean(1)
peaks = []
text = ""
PIN = ""

print("my rec: " + str(myrec))
plt.title("Miau recording")
plt.plot(myrec)
plt.savefig('miau_plot.png')

for i in range(0, len(myrec), 100):
    if(myrec[i] > 0.1):
        peaks.append(i)

# print("peaks: " + str(peaks))

for i in range(len(peaks)-1):
    distance = peaks[i+1] - peaks[i]
    # print("distance: " + str(distance))
    if(distance > 130000):
        if(distance > 180000):
            text += "0"
        else:
            text += "1"

# print("text:" + str(text))

for i in range(0, len(text), 4):
    PIN += str(int(text[i:i+4],2))

print(f"Your PIN is: {PIN}")
```
