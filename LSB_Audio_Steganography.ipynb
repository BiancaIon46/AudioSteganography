```python
import random
def encrypt_data(message):
    ascii_characters = ''.join(chr(i) for i in range(1, 127))
    
    pad = random.randrange(1, 10)
    fb = random.randrange(1, 3)

    # print("fb = " + str(fb) + " pad = " + str(pad))

    fb = int(fb)
    pad = int(pad)

    initial_fb = fb

    encrypted_data = ""

    for i in range(len(message)):
        if fb % 2 == 1:
            encrypted_data += ascii_characters[(ascii_characters.index(message[i]) + pad) % len(ascii_characters)]
            #print((ascii_characters.index(message[i]) + pad) % len(ascii_characters))
            #print("i = " + str(i) + " Encrypted data: " + encrypted_data)
        else:
            new_index = (ascii_characters.index(message[i]) - pad)
            if new_index < 0:
                encrypted_data += ascii_characters[new_index]
            else:
                encrypted_data += ascii_characters[new_index % len(ascii_characters)]
            #print("new index: " + str(new_index))
            #print("i = " + str(i) + " Encrypted data: " + encrypted_data)

        #print(encrypted_data)
        fb += 1

    encrypted_data += str(pad)
    encrypted_data += str(initial_fb)

    return encrypted_data

def decrypt_data(encrypted_data):
    ascii_characters = ''.join(chr(i) for i in range(1, 127))
    #print(ascii_characters)
    print("encrypted data: " + encrypted_data)
    
    fb = encrypted_data[-1:]
    pad = encrypted_data[len(encrypted_data) - 2]

    # print("fb = " + str(fb) + " pad = " + str(pad))

    fb = int(fb)
    pad = int(pad)

    decrypted_data = ""

    for i in range(len(encrypted_data)-2):
        if fb % 2 == 0:
            decrypted_data += ascii_characters[(ascii_characters.index(encrypted_data[i]) + pad) % len(ascii_characters)]
            #print((ascii_characters.index(message[i]) + pad) % len(ascii_characters))
            #print("i = " + str(i) + " Decrypted data: " + decrypted_data)
        else:
            new_index = (ascii_characters.index(encrypted_data[i]) - pad)
            if new_index < 0:
                decrypted_data += ascii_characters[new_index]
            else:
                decrypted_data += ascii_characters[new_index % len(ascii_characters)]
            #print("new index: " + str(new_index))
            #print("i = " + str(i) + " Decrypted data: " + decrypted_data)

        #print(encrypted_data)
        fb += 1


    return decrypted_data
```

```python
def get_binary(message):
    if( type(message) == str ):
        binary_message = ( char.split('b')[1] for char in map(bin,bytearray(message,'utf8')) )
        binary_message_8b = []
        for binary_char in binary_message:
            to_fill = ''.join('0' for l in range(len(binary_char), 8))
            binary_message_8b.append((to_fill + binary_char))
        return ''.join(binary_message_8b)
    elif( type(message) == np.uint8 or type(message) == int  ):
        binary_message = bin(message).split('b')[1]
        to_fill = ''.join('0' for l in range(len(binary_message), 8))
        return to_fill + binary_message
    elif( type(message) == np.ndarray ):
        binary_message_8b = []
        for value in message:
            binary_message = bin(value).split('b')[1]
            to_fill = ''.join('0' for l in range(len(binary_message), 8))
            binary_message_8b.append(to_fill + binary_message)
        return ''.join(binary_message_8b)
```

```python
import tkinter as tk
from tkinter import messagebox, filedialog
from tkinter import *
import wave

class MyGUI:

    def __init__(self):
        self.root = tk.Tk()

        self.root.title("Secret Notes")
        self.root.geometry("800x600")
        self.root.config(bg="#808080")

        self.secretNotes = tk.Label(self.root, text="Secret Notes", fg="white", bg="black", font=('Arial bold', 18))
        self.secretNotes.pack(padx=10)

        self.root.rowconfigure(0, weight = 0)
        self.root.rowconfigure((1, 2, 3, 4, 5, 6), weight = 1)
        
        self.root.columnconfigure(0, weight = 1)
        self.root.columnconfigure(1, weight = 1)

        self.secretNotes.grid(row = 0, columnspan = 2, sticky='we')

        # column 1
        self.encodeLabel = tk.Label(self.root, text="Encode audio file", bg="grey", font=('Arial bold', 14))
        self.encodeLabel.grid(row = 1, column = 0, padx = 2, pady = 2)

        self.chooseEncodeFileButton = tk.Button(self.root, text="Choose encode audio file", fg="white", bg="black", font=('Arial', 12), command=self.chooseAudioFile)
        self.chooseEncodeFileButton.grid(row = 2, column = 0)

        self.insertMessageLabel = tk.Label(self.root, text="Insert message to be encoded:", fg="black", bg="grey", font=('Arial', 12))
        self.insertMessageLabel.grid(row = 4, column = 0)
        
        self.encodeMessageTextbox = tk.Text(self.root, height = 1, width = 25, font=('Arial', 12))
        self.encodeMessageTextbox.grid(row = 5, column = 0)

        self.chooseEncodeFileButton = tk.Button(self.root, text="Encode", fg="white", bg="black", font=('Arial', 12), command=self.encode_message)
        self.chooseEncodeFileButton.grid(row = 6, column = 0)

        # column 2
        self.decodeLabel = tk.Label(self.root, text="Decode audio file", bg="grey", font=('Arial bold', 14))
        self.decodeLabel.grid(row = 1, column = 1, padx = 2, pady = 2)

        self.chooseDecodeFileButton = tk.Button(self.root, text="Choose decode audio file", fg="white", bg="black", font=('Arial', 12), command=self.chooseDecodeAudioFile)
        self.chooseDecodeFileButton.grid(row = 2, column = 1)

        self.chooseDecodeFileButton = tk.Button(self.root, text="Decode", fg="white", bg="black", font=('Arial', 12), command=self.decode_message)
        self.chooseDecodeFileButton.grid(row = 6, column = 1)

        self.filePath = ""
        self.output_file = ""
        self.decode_file = ""
        
        self.root.mainloop()

    def chooseAudioFile(self):
        print("miau")
        self.filePath = filedialog.askopenfilename()
        print(self.filePath)

        self.filePathLabel = tk.Label(self.root, text=self.filePath, fg="black", bg="grey", font=('Arial', 8))
        self.filePathLabel.grid(row = 3, column = 0)

    def chooseDecodeAudioFile(self):
        print("miau")
        self.decode_file = filedialog.askopenfilename()
        print(self.decode_file)

        self.decodeFilePathLabel = tk.Label(self.root, text=self.decode_file, fg="black", bg="grey", font=('Arial', 8))
        self.decodeFilePathLabel.grid(row = 3, column = 1)

    def encode_message(self):
        message = ""
        message = self.encodeMessageTextbox.get("1.0", "end-1c")
        print(message)
        
        if self.filePath == "" or message == "":
            tk.messagebox.showerror(title="Error", message="Fields not completed")
        elif self.filePath[-4:] != ".wav":
            tk.messagebox.showerror(title="Error", message="The file must be .wav format.")
        else:
            self.output_file = self.filePath[:self.filePath.rfind("/") + 1] + "stegano_audio.wav"
            print(self.output_file)
            
            audio = wave.open(self.filePath, mode='rb')
            
            noOfFrames = audio.getnframes()
            byte_frames = audio.readframes(noOfFrames)
            frames = bytearray(list(byte_frames))
            
            encrypted_message = encrypt_data(message)
            
            binary_message = get_binary(encrypted_message)
            
            binary_message += "1111111111111110"
            
            if len(binary_message) > len(frames) * 2:
                raise Exception("The message cannot be encoded due to its lenghts.")

            mask = 252
            index = 0
            for i in range(len(frames)):
                if index < len(binary_message):
                    last_2_bits = binary_message[index:index+2]
                    int_value_to_be_encoded = int(last_2_bits, 2)
                    frames[i] = frames[i] & mask | int_value_to_be_encoded
                    index += 2
                else:
                    break
            
            with wave.open(self.output_file, 'wb') as encoded_audio:
                encoded_audio.setparams(audio.getparams())
                encoded_audio.writeframes(frames)

            tk.messagebox.showinfo(title="Encoding complete", message="Message encoded successfully!\nPath to encoded file: " + self.output_file)
            print("Message encoded")

    def decode_message(self):
        if self.decode_file == "":
            tk.messagebox.showerror(title="Error", message="Fields not completed")
        elif self.decode_file[-4:] != ".wav":
            tk.messagebox.showerror(title="Error", message="The file must be .wav format.")
        else:
            audio = wave.open(self.decode_file, mode='rb')
            
            noOfFrames = audio.getnframes()
            byte_frames = audio.readframes(noOfFrames)
            frames = bytearray(list(byte_frames))
            
            extracted_message = ""
            mask = 3
            for frame in frames:
                extracted_message += format(frame & mask, '02b')
                if extracted_message[-16:] == "1111111111111110":
                    break
            
            decoded_message = ""
            for i in range(0, len(extracted_message), 8):
                byte = extracted_message[i:i+8]
                decoded_message += chr(int(byte, 2))
                if decoded_message[-1] == '\x00':
                    break
            decrypted_message = decrypt_data(decoded_message[:-2])
            print("Decoded message: " + decrypted_message)

            self.showSecretLabel = tk.Label(self.root, text="The secret message is:", fg="black", bg="grey", font=('Arial', 12))
            self.showSecretLabel.grid(row = 4, column = 1)

            self.decodeMessageTextbox = tk.Text(self.root, height = 1, width = 25, font=('Arial', 12))
            self.decodeMessageTextbox.grid(row = 5, column = 1)
            self.decodeMessageTextbox.insert(END, decrypted_message)

            pathToSecretFile = self.filePath[:self.filePath.rfind("/") + 1] + "SecretMessage.txt"
            with open(pathToSecretFile, "w") as textFile:
                textFile.write(decrypted_message)
            
            tk.messagebox.showinfo(title="Encoding complete", message="Message decoded successfully!\nPath to file containing the secret message: " + pathToSecretFile)
        
MyGUI()
```
